```python
#SOFTMAX
import math


logits = [2.0, 1.0, 0.1]


def sigmoid(x):
    return 1 / (1 + math.exp(-x))

sigmoid_output = [sigmoid(x) for x in logits]

def softmax(values):
    exp_values = [math.exp(x) for x in values]
    total = sum(exp_values)
    return [x / total for x in exp_values]

softmax_output = softmax(logits)


print("Logits:", logits)
print("Sigmoid Output:", sigmoid_output)
print("Sum of Sigmoid Outputs:", sum(sigmoid_output))

print("\nSoftmax Output:", softmax_output)
print("Sum of Softmax Outputs:", sum(softmax_output))
```

```python
#Softmax + Cross Entropy Loss
import numpy as np


logits = np.array([2.0, 1.0, 0.1])


exp_logits = np.exp(logits)
softmax_probs = exp_logits / np.sum(exp_logits)
true_class = 0
manual_loss = -np.log(softmax_probs[true_class])

print("Softmax Probabilities:", softmax_probs)
print("Manual Cross Entropy Loss:", manual_loss)
one_hot = np.array([1, 0, 0])
library_loss = -np.sum(one_hot * np.log(softmax_probs))

print("Library Cross Entropy Loss:", library_loss)
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
logits = np.random.randn(5)
exp_logits = np.exp(logits)
softmax_probs = exp_logits / np.sum(exp_logits)

classes = [f"Class {i}" for i in range(len(logits))]

plt.figure()
plt.bar(classes, softmax_probs)
plt.xlabel("Classes")
plt.ylabel("Probability")
plt.title("Softmax Probability Distribution")
plt.show()


print("Logits:", logits)
print("Softmax Probabilities:", softmax_probs)
print("Sum of Probabilities:", np.sum(softmax_probs))
```

```python
import numpy as np


vocab = ["I", "like", "machine", "Translation"]
logits = np.array([2.0, 1.0, 3.0, 0.5])
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

probs = softmax(logits)

print("Vocabulary:", vocab)
print("Logits:", logits)
print("Softmax Probabilities:", probs)

greedy_index = np.argmax(probs)
greedy_word = vocab[greedy_index]
print("\nGreedy Decoding Word:", greedy_word)


sampled_index = np.random.choice(len(vocab), p=probs)
sampled_word = vocab[sampled_index]
print("Random Sampling Word:", sampled_word)
sentence_greedy = "I " + greedy_word
sentence_sampling = "I " + sampled_word

print("\nGenerated Sentence (Greedy):", sentence_greedy)
print("Generated Sentence (Sampling):", sentence_sampling)
```
